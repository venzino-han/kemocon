```python
def print_files_in_dir(root_dir, prefix):
    files = os.listdir(root_dir)
    for file in files:
        path = os.path.join(root_dir, file)
        print(prefix + path)
        if os.path.isdir(path):
            print_files_in_dir(path, prefix + "    ")

print_files_in_dir("data/raw", "")
```

```python
def files_in_dir(root_dir):
    fileList =[]
    files = os.listdir(root_dir)
    for file in files:
        path = os.path.join(root_dir, file)
        if isfile(path):
            fileList.append(path)
        if os.path.isdir(path):
            fileList += files_in_dir(path)
            
    return fileList

files_in_dir("data\\raw\\neurosky_polar_data")
```

```python
anno_types =['self','partner','external','agg_ex']
anno_file_dic={}

for t in anno_types:
    u_anno_path = os.path.join(anno,t)
    anno_data = files_in_dir(u_anno_path)
    anno_file_dic[t]=anno_data
```

```python
# check point 제외 
# P+uid 파일만 저장 

u_dic = {}
for t in anno_types:
    files = anno_file_dic[t]
    for f in files:
        f = f.split('\\')[-1]
        p = f.split('.')[0]
        if p == 'P'+uid and 'checkpoint' not in f:
            if t in u_dic:
                u_dic[t].append(f)
            else:
                u_dic[t]=[f]
```
