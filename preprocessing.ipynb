```python
import os
import pandas as pd
import numpy as np
from os.path import *
```

```python
root = 'data'
neuro = os.path.join(root, 'raw','neuro')
anno = os.path.join(root, 'raw', 'anno')
e4 = os.path.join(root, 'raw', 'e4')
```

```python
users = list(range(1,33))
```

```python
def files_in_dir(root_dir):
    fileList =[]
    files = os.listdir(root_dir)
    for file in files:
        path = os.path.join(root_dir, file)
        if isfile(path):
            fileList.append(path)
        if os.path.isdir(path):
            fileList += files_in_dir(path)
            
    fileList = [ x for x in fileList if ('checkpoint' not in x) ]
            
    return fileList
```

```python
def get_labels(uids):
    anno_types =['self','partner','external','agg_ex']
    anno_file_dic={}

    for t in anno_types:
        u_anno_path = os.path.join(anno,t)
        anno_data = files_in_dir(u_anno_path)
        anno_file_dic[t]=anno_data

    u_dic = {}
    for t in anno_types:
        files = anno_file_dic[t]
        for f in files:
            f = f.split('\\')[-1]
            p = f.split('.')[0]
            if p == 'P'+uid and 'checkpoint' not in f:
                if t in u_dic:
                    u_dic[t].append(f)
                else:
                    u_dic[t]=[f]
                    
    return u_dic
```

```python
## data complete check 
dataCompDic={}

for uid in users:
    comp ={}
    for t in [neuro, e4]:
        key = t.split('\\')[-1]
        path = os.path.join(t,str(uid))
        try:
            files = files_in_dir(path)
            comp[key] = [ x.split('\\')[-1] for x in files]
        except:
            comp[key] = []
        
    dataCompDic[uid] = comp
```

```python
#Concat all df 
#recursivly

def concatAll(df1, dfs):
    if dfs == []:
        return df1
    
    df2 = dfs.pop()
    
    df1 = pd.merge(df1,df2, how='outer', on='timestamp')
    return concatAll(df1,dfs)
```

```python
dfs = list(neuro_dfs.values())
df1 = dfs.pop()

neuro_df = concatAll(df1, dfs)
```
